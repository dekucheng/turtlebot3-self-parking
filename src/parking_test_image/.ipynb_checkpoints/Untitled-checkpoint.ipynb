```python
top_x = 85
top_y = 85
bottom_x = 290
bottom_y = 235
l_u = [320 - top_x, 360 - 2*top_y]
r_u = [320 + top_x, 360 - 2*top_y]
r_b = [320 + bottom_x, 240 + bottom_y]
l_b = [320 - bottom_x, 240 + bottom_y]
m_u = [320, 360 - 2*top_y]
m_b = [320, 240 + bottom_y]
m_m = [320, (600 - 2*top_y+bottom_y)/2]
l_m = [(640-top_x-bottom_x)/2, (600-2*top_y+bottom_y)/2]
r_m = [(640+top_x+bottom_x)/2, (600-2*top_y+bottom_y)/2]

pts_src = np.array([l_u,m_u,r_u,r_m,r_b,m_b,l_b,l_m, m_m])

# selecting 4 points from image that will be transformed
pts_dst = np.array([[200, 0], [500,0], [800, 0], [800,300], [800, 600], [400,600], [200, 600], [200,300], [500,300]])

# finding homography matrix
h, status = cv2.findHomography(pts_src, pts_dst,method=0)
h1 = h[0][0]
h2 = h[0][1]
h3 = h[0][2]
h4 = h[1][0]
h5 = h[1][1]
h6 = h[1][2]
y = ((h1*y2+h4*x2)-h3*h4-h1*h6)/(h1*h5+h2*h4)
x = (x2-h3-h2*y)/h1
print(x,y)
cv_image_homography = cv2.warpPerspective(image, h, (1000, 600))
```
